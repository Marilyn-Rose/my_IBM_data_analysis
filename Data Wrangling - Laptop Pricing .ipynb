```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import requests
```

```python
def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(response.content)

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
file_name = "lapotops.csv"
download(url, file_name)
```

```python
df = pd.read_csv(file_name, header = 0 )
```

```python
df.replace("?", np.nan, inplace = True)
```

```python
# Using the method 'df.isnull()' to check if there are missing values in the data

missing_data = df.isnull()
missing_data
```

```python
# a for loop that gives the number of missing values currently available in the every coloumn:

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
# the columns and the number of missing values available are:
# 'Screen_Size_cm' : 4
# 'Weight_kg' : 5
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Considering the 'Screen_Size_cm' column, it's values are categorical in nature. So its best t replace the missing values with the most frequent value of the attribute.
# i.e. replacing it with its mode

common_screen_size = df['Screen_Size_cm'].value_counts().idxmax()
print("The most common screen size in cm is: ", common_screen_size)

#Replace the missing values with the gotten mode
df['Screen_Size_cm'].replace(np.nan, common_screen_size, inplace = True)
```

```python
# In order to deal with the missing values, the average of the entire values is used to replace the missing value
avg_weight = df['Weight_kg'].mean(axis=0)
print("Average weight in kg is: ", avg_weight)

df['Weight_kg'].replace(np.nan, avg_weight, inplace = True)
```

```python
# Checking again to see if the data still has missing data :
missing_data2 = df.isnull()

for column in missing_data2.columns.values.tolist():
    print(column)
    print (missing_data2[column].value_counts())
    print("")
```

```python
# update the 'Screen_size_cm' column so that it's values are rounded up to the nearest 2 decimal places

df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']], 2)
df.head()
```

```python
# DATA STANDARDIZATION
# the value Screen_size standard unit is inches. Similarly, weight of the laptop is needed to be in pounds.
# use the conversion below to modify the columns
# 1 inch = 2.54 cm
# 1 kg = 2.205 pounds

df['Screen_Size_inches'] = df['Screen_Size_cm']/2.54
df['Weight_pounds'] = df['Weight_kg']*2.205
```

```python
df.head()
```

```python
# DATA NORMALIZATIN
# Let's normalize the 'CPU_frequency' attribute with respect to maximum value available 
df["CPU_frequency"] = df["CPU_frequency"]/df["CPU_frequency"].max()
```

```python
# BINNING
# Creat 3 bins for the attribute "Price". Name them Low, Medium and High 
# The attribute will be named "Price-binned"

bins = np.linspace(min(df["Price"]), max(df["Price"]), 4)
group_names = ["Low", "Medium", "High"]
df["Price-binned"] = pd.cut(df["Price"], bins, labels = group_names, include_lowest = True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.bar(group_names, df["Price-binned"].value_counts())

plt.pyplot.xlabel("Price")
plt.pyplot.ylabel("count")
plt.pyplot.title("Price bins")
```

```python
# INDICATIOR VARIABLES
dummy_variable = pd.get_dummies(df['Screen'])
dummy_variable.head()
```

```python
dummy_variable.rename(columns = {'IPS Panel': 'Screen-IPS_panel', 'Full HD': 'Screen-Full_HD'}, inplace = True)
df = pd.concat([df, dummy_variable], axis = 1)

# drop original column 'Screen'
df.drop("Screen", axis = 1, inplace = True)
```
