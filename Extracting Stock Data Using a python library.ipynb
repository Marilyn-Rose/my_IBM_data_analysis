```python
import yfinance as yf
import pandas as pd
```

```python
#Using the yfinance library to extract stock data
# using the 'Ticker' module to create an object that would allow us to 
#access functions to extract data.

# The company is Apple and the ticker symbol is AAPL

apple = yf.Ticker("AAPL")
```

```python
import requests
import json
response = requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json")
```

```python
apple_info = response.json()
apple_info
```

```python
apple_info['country']
```

```python
apple_share_price_data = apple.history(period='100y')
```

```python
apple_share_price_data
```

```python
apple_share_price_data.reset_index(inplace = True)
```

```python
apple_share_price_data.plot(x = "Date", y = "Open")
```
