```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
print('Matplotlib version: ', mpl.__version__)
```

```python
import requests
from io import StringIO

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv"
response = requests.get(URL)
csv_data = StringIO(response.text)

df_can = pd.read_csv(csv_data)
print('Data read into a pandas dataframe!')
```

```python
df_can.head()
```

```python
print (df_can.shape)
```

```python
df_can.set_index('Country', inplace = True)
df_can.head()
```

```python
print('data dimensions:', df_can.shape)
```

```python
years = list(map(str, range(1980, 2014)))
```

# Line Plot

Create a line plot to visualize the immigrants ( to Canada) trend during 1980 to 2013.


```python
# creating df with only years columns from 1980 - 2013
df_line = df_can[years]

# Applying sum to get total immigrants year-wise
total_immigrants = df_line.sum()
total_immigrants
```

```python
# Create figure and axes
fig, ax = plt.subplots()

# plot the line
ax.plot(total_immigrants)

# Setting up the Title
ax.set_title('Immigrants between 1980 to 2013')

# Setting up the labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')

# Display the plot
plt.show()
```

```python
# Create figure and axes
fig, ax = plt.subplots()

# changing the index type to integer
total_immigrants.index = total_immigrants.index.map(int)

# plot the line
ax.plot(total_immigrants)

# Setting up the Title
ax.set_title('Immigrants between 1980 to 2013')

# Setting up the labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')

# Display the plot
plt.show()
```

#### Customize the above plot's appearance

```python
# Create figure and axes
fig, ax = plt.subplots()

# changing the index type to integer
total_immigrants.index = total_immigrants.index.map(int)

# Customizing the appearance of plot
ax.plot(total_immigrants,
       marker = 's',
       markersize = 5,
       color = 'green',
       linestyle = 'dotted')

# Setting up the Title
ax.set_title('Immigrants between 1980 to 2013')

# Setting up the labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')
ax.legend(['Immigrants'])

# Display the plot
plt.show()
```

#### Plot a line graph of immigration from Haiti

```python
# creating data for plotting:
df_can.reset_index(inplace = True)
haiti = df_can[df_can['Country'] == 'Haiti']


# creating haiti with only years columns from 1980 - 2013 
# and transpoding to get the result as a series
haiti = haiti[years].T

# converting the index to type interger
haiti.index = haiti.index.map(int)

```

```python
# plotting the line plot on the data
fig, ax = plt.subplots()
ax.plot(haiti)

#Title
ax.set_title('Immigrants from Haiti between 1980 to 2013')

# Setting up the labels
ax.set_xlabel('Years')
ax.set_ylabel('Number of Immigrants')

plt.legend(['Immigrants'])

plt.show()
```

##### We can clearly notice how number of immigrants from Haiti spiked up from 2010 as Canada stepped up its efforts to accept refugees from Haiti.
##### Let's annotate this spike in the plto by using the *ax.annotate()* method.

```python
# plotting the line plot on the data
fig, ax = plt.subplots()
ax.plot(haiti)

#Title
ax.set_title('Immigrants from Haiti between 1980 to 2013')

# Setting up the labels
ax.set_xlabel('Years')
ax.set_ylabel('Number of Immigrants')

plt.legend(['Immigrants'])

ax.set_xticks(list(range(1980,2015,5)))

ax.annotate('2010 Earthquake', xy = (2000, 6000))

plt.show()
```

# Scatter Plot

Create a *Scatter plot* to visualize the immigrants (to Canada) trend during 1980 to 2013.
We  need the Total of year-wise immigrants

Create a new datafrane for only columns containing the years then, we will apply the sum() on the dataframe.

Create a scatter plot direclty on ax by calling the plot function()

```python
# Create figure and axes
fig, ax = plt.subplots(figsize =(8, 4))

# Customizing scatter plot 
ax.scatter(years, total_immigrants,
           marker = 'o', # setting up the markers
           s = 20, # setting up the size of the marker
           color = 'darkblue'
          )

plt.title("Immigrants between 1980 to 2013")

plt.xlabel('Years')
plt.ylabel('Total Immigrants')

plt.grid(True)  # including grid

ax.legend(['Immigrants'], loc = 'upper center')

plt.show()
```

# Bar Plot

Create a bar plot to visualize the top 5 countries that contributed the most immigrants to Canada from 1980 to 2013.

Apply sort_values function on the 'Total' column of our data

Create a new dataframe for only columns containing the years then, apply sum() on the dataframe and create a separatedataframe for top five countries.

# Histogram

Find the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?

```python
df_country = df_can.groupby(['Country'])['2013'].sum().reset_index()

# Create figure and axes
fig, ax = plt.subplots(figsize=(10,4))
ax.hist(df_country['2013'])
ax.set_title('New Immigrants in 2013')
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Countries')
ax.legend(['Immigrants'])

# Display the plot
plt.show()
```

###### 
The above plot does not match with the bars


```python

# Plot the bar
fig, ax = plt.subplots(figsize=(10,4))
count = ax.hist(df_country['2013'])


ax.set_title('New Immigrants in 2013')
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Countries')
ax.set_xticks(list(map(int, count[1])))
ax.legend(['Immigrants'])

# Display the plot
plt.show()
```

#### We can also plot multiple histograms on the same plot. 

#### What is the immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013?

```python
df = df_can.groupby(['Country'])[years].sum()
df_dns = df.loc[['Denmark', 'Norway', 'Sweden'], years]
df_dns = df_dns.T
df_dns
```

```python
# Create figure and axes
fig, ax = plt.subplots(figsize = (10,4))
ax.hist(df_dns)
ax.set_title('Immigration from Denmark, Norway, and Sweden from 1980 to 2013')
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Years')
ax.legend(['Denmark', 'Norway', 'Sweden'])

plt.show()
```

#### What is the immigration distribution for China and India for years 2000 to 2013?

```python
df = df_can.groupby(['Country'])[years].sum()
years1 = list(map(str, range(2000, 2014)))
df_CI = df.loc[['China', 'India'], years1]
df_CI = df_CI.T
df_CI
```

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.hist(df_CI)
ax.set_title('Immigration from China and India from 2000 to 2013')
ax.set_ylabel('Number of Years')
ax.legend(['China', 'India'])

plt.show()
```

#  Pie Chart

Create a pie chart representing the 'Total Immigrants' for the year 1980 to 1985

```python
fig, ax = plt.subplots()

ax.pie(total_immigrants[0:5], labels = years[0:5],
       colors = ['gold', 'blue', 'lightgreen', 'coral', 'cyan'],
       autopct = '%1.1f%%', explode = [0,0,0,0,0.1]) # using explode to highlight the lowest

ax.set_aspect('equal')  # Ensure pie is drawn as a circle

plt.title('Distribution of Immigrants from 1980 to 1985')

plt.show()
       
```

#### Create a pie chart representing the total immigrants proportion for each continent

First, group the data over continents and get the sum on total. Then pass this data to the pie function.

```python
# Creating data for plotting pie
df_con = df_can.groupby('Continent')['Total'].sum().reset_index()
label = list(df_con.Continent)
label[3] = 'LAC'
label[4] = 'NA'
df_con
```

```python
fig, ax = plt.subplots(figsize =(10,4))

ax.pie(df_con['Total'], 
      colors = ['gold', 'blue', 'lightgreen', 'coral', 'cyan', 'red'],
       autopct = '%1.1f%%',
       pctdistance = 1.25
      )

ax.set_aspect('equal')

plt.title('Contintent-wise distribution of immigrants')
ax.legend(label, bbox_to_anchor = (1, 0, 0.5, 1))

plt.show()
```

# Sub-plotting

Create a line and scatter plot in one row plt.subplots()

Assign the 'Sharey' parameter as True in the code below. Also notice the use of 'suptitle'

```python
# Create a figure with two axes in a row

fig, axs = plt.subplots(1, 2, sharey = True)

# Plotting in first axes - the left one
axs[0].plot(total_immigrants)
axs[0].set_title('Line plot on immigrants')

# Plotting in first axes - the left one
axs[1].scatter(total_immigrants.index, total_immigrants)
axs[1].set_title('Scatter plot on immigrants')

axs[0].set_ylabel('Number of Immigrants')

# Adding a Title for the Overall Figure
fig.suptitle('Subplotting Example', fontsize = 15)

# Adjust spacing between subplots
fig.tight_layout()

# Show the figure
plt.show()
```

###### 
Implementing the subplotting with **add_subplot()** below:

```python
# Create a figure with Four axes - two rows, two columns
fig = plt.figure(figsize=(8,4))

# Add the first subplot (top-left)
axs1 = fig.add_subplot(1, 2, 1)
# Plotting in first axes - the left one
axs1.plot(total_immigrants)
axs1.set_title('Line plot on immigrants')

# Add the second subplot (top-right)
axs2 = fig.add_subplot(1, 2, 2)
#Plotting in second axes - the right one
axs2.barh(total_immigrants.index, total_immigrants) #Notice the use of 'barh' for creating horizontal bar plot
axs2.set_title("Bar plot on immigrants")
            
#Adding a Title for the Overall Figure
fig.suptitle('Subplotting Example', fontsize=15)

# Adjust spacing between subplots
fig.tight_layout()


plt.show()
```
