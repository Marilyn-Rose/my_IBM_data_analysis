```python
import pandas as pd
import numpy as np
```

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx'
df_can = pd.read_excel(url,
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)

print('Data read into a pandas dataframe!')
```

```python

# NOTE: in pandas axis = 0 represents rows(default) and axis = 1 represents column
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)
df_can.head(2)
```

```python

# Renaming the columns for better understanding:
df_can.rename(columns= {'OdName': 'Country', 'AreaName':'Continent', 'RegName': 'Region'}, inplace = True)
df_can.head(2)
```

```python

numeric_df = df_can.select_dtypes(include=[np.number]) # selects only the numeric columns from the DataFrame.

# Add a 'Total' Column that sums up the total immigrants by country over the entire period 1980 - 2013:
df_can['Total'] = numeric_df.sum(axis=1)
df_can['Total']
```

```python
# 1. Create the condition boolean series
condition = df_can['Continent'] == 'Asia'
print(condition)
```

```python

# Pass the conditoin into the dataFrame:
df_can[condition]
```

```python

# Passing multiple criteria in the same line
# Let's filter for AreaName = Asia and RegName = Southern Asia

df_can[(df_can['Continent'] == 'Asia') & (df_can['Region'] == 'Southern Asia')]
```

```python

# Exercise: Fetch the data where AreaName is 'Africa' and RegName is 'Southern Africa'.
df_can[(df_can['Continent'] == 'Africa') & (df_can['Region'] == 'Southern Africa')]
```

```python

# Sorting our dataframe df_can on 'Total' column, in descending order ot find out the top
# 5 countries that contrivuted the most to immigration to Canda.

df_can.sort_values(by = 'Total', ascending = False, axis = 0, inplace = True)
top_5 = df_can.head(5)
top_5
```
